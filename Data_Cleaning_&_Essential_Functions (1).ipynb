(1)what is data cleaning, and why is it important in data analysis?

What are the potential consequences of analyzing unclean or messy data?

ans (1) 1. What is data cleaning
Answer: Data cleaning is the process of identifying and correcting errors, inconsistencies, and inaccuracies in a dataset to improve its quality.
Step 1: Importance in data analysis
Data cleaning is important because accurate and consistent data leads to reliable analysis and sound decision-making.
Step 2: Potential consequences of analyzing unclean data
Analyzing unclean or messy data can lead to:
Incorrect conclusions and flawed insights.
Misguided business decisions.
Wasted time and resources.
Difficulty in integrating data from multiple sources.
Step 3: Common steps involved in cleaning and organizing data
Common steps include:
Handling Missing Data: Identifying and addressing gaps in the dataset (e.g., imputation or removal).
Removing Duplicates: Eliminating redundant entries.
Correcting Errors: Fixing typos, structural errors, and inconsistencies.
Standardizing Formats: Ensuring uniformity in data types, dates, and units.
Validating Data: Checking data against predefined rules or constraint

3. Explain the use of text functions



Answer: Text functions are used in data cleaning to manipulate and format text data.
TRIM: Removes leading, trailing, and extra spaces within text.
LEFT: Extracts a specified number of characters from the beginning (left side) of a text string.
RIGHT: Extracts a specified number of characters from the end (right side) of a text string.
MID: Extracts a substring from the middle of a text string, starting at a specified position and for a specified length.
CONCAT (or CONCATENATE): Joins two or more text strings into one.

8. Understanding TODAY() and NOW()
a. Difference between TODAY() and NOW()



Answer: (a) TODAY() returns the current date only, while NOW() returns both the current date and time.
TODAY() is useful when you only need the date, for example, to calculate age or the number of days until a future date.
NOW() is useful when you need both date and time, for example, to timestamp when a task was completed.
b. Calculate days remaining until a future date
Step 1: Write the formula
If cell A1 contains the date 2025-06-10, the formula to determine how many days are left until that date using TODAY() is to subtract today's date from the future date:
=A1 - TODAY()
Answer:
=A1 - TODAY()
c. Display current date and time in a specific format
Step 1: Write the formula
To display the current date and time in the format MM/DD/YYYY HH:MM AM/PM using the NOW() function, you can use the TEXT function to format the output:
=TEXT(NOW(), "MM/DD/YYYY HH:MM AM/PM")
Answer:
=TEXT(NOW(), "MM/DD/YYYY HH:MM AM/PM")
d. Behavior of =TODAY() when reopened
Answer: (d) The cell value will update to the new current date.
TODAY() is a volatile function, meaning it recalculates every time the worksheet is opened or any change is made to the workbook.
e. Keyboard shortcut for a static date
Answer: (e) Ctrl + ;
This shortcut enters the current date as a static value that does not change when the workbook is reopened.
